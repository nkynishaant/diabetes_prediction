### Disease predicter
## Diabetic predeiction

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm 
from sklearn.metrics import accuracy_score 
```

```python
# DATA PREPROCESSING
```

```python
data=pd.read_csv('diabetes.csv')
data
```

```python
data.head() # first 5 rows of the dataset
```

```python
#no.of rows and columns in this data set
data.shape
```

```python
#getting statistical measure of data
data.describe()
```

```python
data['Outcome'].value_counts() #0--> non-diabetic
                               #1--> diabetic
```

```python
data.groupby('Outcome').mean()
```

```python
# seperating the data and labels
X=data.drop(columns='Outcome',axis=1)
Y=data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
#data standardization -- to fit and transform the data in a particular range so model will do better prediction 
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=data['Outcome']
```

### train test split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#stratify is used for spliting the diabetic and non-dia in the same proportion
#random state is use to always split the data in the same way 
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

### Training the model

```python
classifier = svm.SVC(kernel='linear')#kernal is another parameter equals to linear model
```

```python
classifier.fit(X_train,Y_train) #training the supporrt vector classifier
```

### model evaluation
### accuracy score

```python
#accuracy score of the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
                                     
```

```python
print('Accuracy score of the training data :',training_data_accuracy)
```

```python
#accuracy score of the test data
X_test_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
                                     
```

```python
print('Accuracy of the Test data :',testing_data_accuracy)
```

## Making a predictive sysytem

```python
input_data=(90,68,42,0,38.2,0.503,27,0)
#changing inputdata to numpy array bcause the processing is more easy and efficient
input_data_as_numpy_array=np.asarray(input_data)#asarray is use to tranform list into numpyarray
#reshape the array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)#this is the diameter we giving to tell the model that we are predicting only one instance
#standardise the data 
std_data=scaler.transform(input_data_reshaped)
print(std_data)
prediction=classifier.predict(std_data)
print(prediction)

if (prediction[0]==1):#[0]this represents the first value in prediction
    print('The person is diabetic')
else:
    print('The person is non diabetic')
```
